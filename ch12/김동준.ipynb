```python
# 다음은 소상공인 시장 진흥곤단에서 제공하는 2년간(2017.10 ~ 2019.09)국내의 카페 업소 수 데이터를 정리한 CSV 파일이다. 물음에 답하시오. (1번 문제)

# 1. Matpoltlib을 이용하여 월별 카페의 개수를 나타내는 간단한 막대 그래프(Bar Graph)를 그리는 프로그램을 작성하시오, 
# 단, 막대 그래프의 x 축 간격을 3달로 설정한다. 
```

```python
import csv

f = open('cafe_2year.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)


x = []
y = []
for line in lines:
    x.append(line[0])
    y.append(line[1])
    #print(line)
f.close()

nx = []
ny = []
for i in range(len(x)):
    if i % 3 == 0:
        nx.append(x[i])
        ny.append(y[i])
        
del (nx[0])
del (ny[0])

plt.title('전국의 월별 카페 수')
plt.bar(nx,ny)
plt.ylabel('카페수')
plt.show()

```

```python
# 5. 4번 문제에서 사용한 doctor2.csv 파일을 읽어 서울과 6개 광역시의 종합병원의
# 일반의, 인턴, 레지던트, 전문의 의사 수를 선 그래프로 나타내는 프로그램을 작성하시오. 

# 4. 3번 문제에서 저장한 doctor2.csv 파일을 읽어들여 서울과 6개 광역시의 종합 병원 전문의 의사 수를
# 산포 그래프(Scatter Plot)로 보여주는 프로그램을 작성하시오.  
import matplotlib.pyplot as plt
import numpy as np
import math as math
from matplotlib import rc

# Matplotlib 한글처리 코드
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font', family='Malgun Gothic')
%matplotlib inline

f = open('doctor2.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)

header = next(lines)    #분리한 헤더는 legend로 넣기
#print(header)


x = []
y1 = []   #일반의
y2 = []   #인턴 
y3 = []   #레지던트
y4 = []   #전문의

for line in lines:
    x.append(line[0])
    y1.append(line[1])
    y2.append(line[2])
    y3.append(line[3])
    y4.append(line[4])
#     print(line)
    
intList1 = [int(x) for x in y1]
intList2 = [int(x) for x in y2]
intList3 = [int(x) for x in y3]
intList4 = [int(x) for x in y4]

plt.ylim(-500,10000)
plt.yticks([0,2000,4000,6000,8000])

plt.plot(x, intList1, label = '일반의', color = 'red', linestyle = '-', marker = 'o' )
plt.plot(x, intList2, label = '인턴', color = 'blue', linestyle = '--', marker = 'x' )
plt.plot(x, intList3, label = '레지던트', color = 'yellow', linestyle = '-.', marker = 's' )
plt.plot(x, intList4, label = '전문의', color = 'magenta', linestyle = '-.', marker = 'd' )

plt.legend(loc='best')
plt.xlabel('시')       
plt.ylabel('의사 수')  
plt.title("서울 및 광역시 종합병원 의사 수")

plt.show()
```

```python
# 6. 4번 문제에서 사용한 doctor2.csv 파일을 읽어 서울 종합병원의 의사 수 분포를 의사 유형별로 
# 나타내는 파이 그래프를 그리는 프로그램을 작성하시오.

import matplotlib.pyplot as plt
import numpy as np
import math as math
from matplotlib import rc

# Matplotlib 한글처리 코드
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['axes.unicode_minus'] = False
mpl.rc('font', family='Malgun Gothic')
%matplotlib inline

f = open('doctor2.csv', 'r', encoding = 'utf-8')
lines = csv.reader(f)

header = next(lines)    #분리한 헤더는 legend로 넣기
#print(header)

x = []

sum = 0
lv = ['일반의','인턴','레지던트','전문의']
for line in lines:
    
    if line[0] == '서울' :
        x = line


header = x[0]
del (x[0])

intList1 = [int(_x) for _x in x]

sum = 0
for i in intList1:
    sum += i 

avgList1 = []
for i in range(4):
    x = intList1[i] / sum * 100
    
    avgList1.append(round(x,2))
#print(avgList1)

portion = []
for i in avgList1:
    portion.append(i)

plt.pie(portion, explode = (0 ,0.2, 0, 0), labels = lv, autopct = '%1.1f%%', shadow = False, startangle = 90)

plt.title('서울 종합병원 의사 유형별 분포')

```
