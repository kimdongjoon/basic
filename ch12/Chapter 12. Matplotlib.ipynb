```python
xdata = ['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]

plt.plot(xdata,ydata1, label = '국어')
plt.plot(xdata,ydata2, label = '영어')

plt.legend(loc = 'upper center')

plt.title('세명 학생의 국어, 영어성적')
plt.show()
```

```python
# 12 - 5 Line Style
xdata = ['안지영','홍지수','황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
ydata3 = [85,97,78]
ydata4 = [92,88,82]

plt.plot(xdata,ydata1, label = '국어', color = 'red', linestyle = '-', marker = 'o')
plt.plot(xdata,ydata2, label = '영어', color = '#00ffff', linestyle = '--', marker = 'x')
plt.plot(xdata,ydata3, label = '수학', color = 'magenta', linestyle = '-.', marker = 's')
plt.plot(xdata,ydata4, label = '사회', color = '#444444', linestyle = ':', marker = 'd')

plt.title('세명의 학생의 네 과목 성적')
plt.legend(loc = 'best')

plt.show()
```

```python
# 12- 6
import matplotlib.pyplot as plt
from matplotlib import rc

xdata = list(range(30))
ydata = []

for x in xdata :
    y = 2 * x
    ydata.append(y)
    
plt.plot(xdata,ydata,color= 'magenta', linewidth = 1)
plt.title('y=2x')
plt.xlim(-1,30)
plt.ylim(-2,60)
plt.xticks(list(range(0,35,2)))
plt.yticks(list(range(0,70,5)))

plt.grid(True)
plt.show()
```

```python
# 12- 6
import matplotlib.pyplot as plt
import numpy as np
import math as math
from matplotlib import rc

xdata = np.linspace(0, (2*np.pi), 500)
ydata = np.sin(xdata)
cdata = np.cos(xdata)

plt.xticks(np.arange(0,2*np.pi+0.01 , np.pi/4))


plt.plot(xdata, ydata, color = 'red', label = 'sin' ,linewidth = 1 ,linestyle = '-')
plt.plot(xdata, cdata, color = 'blue', label = 'cos',linewidth = 1 ,linestyle = '-.')
plt.legend(loc = 'best')
plt.show()
```

```python
import numpy as np
import matplotlib.pylab as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
    
#x = np.arange(-3.0, 3.0, 0.1)

x = np.linspace(-3,3,601)
#y =1 / (1+ np.exp(-x))          #Sigmoid 함수
y = sigmoid(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.title('시그모이드 그래프')
plt.axhline(y= 0, color = 'k', linewidth = 0.5)   #수평선(X축)
plt.axhline(y= 1, color = 'k', linewidth = 0.5)
plt.axvline(x= 0, color = 'k', linewidth = 0.5)   #수직선(Y축)

plt.show()
```
