```python
df = pd.read_csv('songs.csv' 
                ,names = ['song','gid','year','composer','lyrics'], encoding = 'CP949')

df
```

```python
songs = []
tmp = []
for i in range(len(df)):
    tmp = i, df['song'][i] , df['gid'][i] , df['year'][i] , df['composer'][i], df['lyrics'][i]
    songs.append(tmp)
```

```python
del songs[0]
```

```python
sql = "insert into song(song_id, song_name,gid,year,composer, lyrics) values(?,?,?,?,?,?)"
for song in songs:
    cur.execute(sql, song)
conn.commit()
```

```python
'''
위 두개의 테이블을 조인한 결과를 가지고 다음의 필드를 갖는 
데이터 프레임을 만드시오.
그룹 이름, 구성원 수, 데뷔 일자, 노래 이름, 발표 년도
'''
```

```python
sql = "SELECT a.name, a.member_no, a.debut , b.song_name, b.lyrics FROM singer as a inner JOIN song as b ON a.id = b.gid"
cur.execute(sql)
results = cur.fetchall()
```

```python
df = pd.DataFrame(results)
df.rename(columns = {0 : '그룹 이름 ', 1 : '구성원 수' , 2 : '데뷔 일자', 3 : '노래 이름' , 4 : '노래내용'} ,inplace = True)

```
